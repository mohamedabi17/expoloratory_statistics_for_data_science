```python

df_transposed = df.transpose()
df_transposed.columns = ['Nombre_Employes']

df_transposed.index = df_transposed.index.str.strip(']').str.split(',').map(lambda x: (float(x[0]) + float(x[1])) / 2)
 #95 des valeurs centrales
percentile_2_5 = np.percentile(np.repeat(df_transposed.index, df_transposed['Nombre_Employes']), 2.5)
percentile_97_5 = np.percentile(np.repeat(df_transposed.index, df_transposed['Nombre_Employes']), 97.5)
print(f"L'intervalle contenant 95% des valeurs centrales est [{percentile_2_5}, {percentile_97_5}]")
```

```python

median = np.median(np.repeat(df_transposed.index, df_transposed['Nombre_Employes']))
mode = df_transposed.idxmax()['Nombre_Employes']
mean = np.average(np.repeat(df_transposed.index, df_transposed['Nombre_Employes']))

print(f"Médiane : {median}")
print(f"Mode : {mode}")
print(f"Moyenne : {mean}")
```

```python

# Comparaison pour déduire le type de la distribution
if median == mode == mean:
    print("La distribution est symétrique et centrée.")
elif median < mode < mean:
    print("La distribution est asymétrique à droite (positivement asymétrique).")
elif median > mode > mean:
    print("La distribution est asymétrique à gauche (négativement asymétrique).")
else:
    print("La distribution est complexe.")
```
